```python
def CircuitReversal(qc):

    f = open("circuit_source_"+str(datetime.datetime.now())+".py", "x")
    
    f.write("# thanks ryan, its been fun. \n")
    f.write("# from qiskit import QuantumRegister, QuantumCircuit, opflow \n")
    f.write("# import numpy \n")

    
    depth = qc.depth()
    
    
    f.write("new_qreg_q = QuantumRegister("+str(depth)+", 'q') \n" 
            "new_qcircuit = QuantumCircuit(new_qreg_q) \n \n")
    
    
    for i in qc.data:
        
        #Taking attributes from the instructions: name of the gate and its parameters.
        name = i.operation.name
        params = i.operation.params
        
        #Setting params to 0 
        param1 = 0
        param2 = 0 
        param3 = 0
        param4 = 0
        
        
        
        qubit_operand_1 = i.qubits[0].index
        
        #It is 'trying' below in case 2nd and 3rd qubit are not defined
        try:
            qubit_operand_2 = i.qubits[1].index
        except:
            pass
        
        try:
            qubit_operand_3 = i.qubits[2].index
        except:
            pass
        
        try:
            param1 = params[0]
            param2 = params[1]
            param3 = params[2]
            param4 = params[3]
        except:
            pass

                        
        #Grouping add gate methods according to qubit and parameter inputs. More work needs to be done re: thinking
        #About speed/efficiency
        
        
        #could possibly use Match-Case (added with python 3.10) with objects and its attributes - may be cleaner + faster
        
        #F-strings added with python 3.6 [supported?]
        if name in ['h', 'sxdg', 'i', 'sx', 'x', 'y', 's', 't', 'sdg', 'tdg', 'z']:
            gate_command = f"new_qcircuit.{name}({qubit_operand_1})\n"
            f.write(gate_command)
            
        elif name in ['cx', 'ecr', 'csx', 'swap', 'dcx', 'ch', 'cy', 'cz', 'cs', 'csdg', 'iswap']:
            gate_command = f"new_qcircuit.{name}({qubit_operand_1}, {qubit_operand_2})\n"
            f.write(gate_command)
            
        elif name in ['ryy', 'crx', 'cry', 'crz', 'cp', 'rxx', 'rzz', 'rzx']:
            gate_command = f"new_qcircuit.{name}({param1}, {qubit_operand_1}, {qubit_operand_2})\n"
            f.write(gate_command)
                 
        elif name in ['ccx', 'cswap', 'ccz', 'rccx']:
            gate_command = f"new_qcircuit.{name}({qubit_operand_1}, {qubit_operand_2}, {qubit_operand_3})\n"
            f.write(gate_command)
    
        elif name in ['rz', 'p', 'rx', 'ry'] :
            gate_command = f"new_qcircuit.{name}({param1}, {qubit_operand_1})\n"
            f.write(gate_command)
            
        elif name == 'u1':
            gate_command = f"new_qcircuit.u(0, 0, {param1}, {qubit_operand_1})\n"
            f.write(gate_command)
            
        elif name == 'u2':
            gate_command = f"new_qcircuit.u((numpy.pi/2), {param1}, {param2}, {qubit_operand_1})\n"
            f.write(gate_command)
            
        elif name in ['u', 'u3']:
            gate_command = f"new_qcircuit.u({param1}, {param2}, {param3}, {qubit_operand_1})\n"
            f.write(gate_command)
            
        elif name == 'r':
            gate_command = f"new_qcircuit.{name}({param1}, {param2}, {qubit_operand_1})\n"
            f.write(gate_command)
        
        elif name == 'cu3':
        #there should be 3 parameters for cu3 - 4 listed here
            gate_command = f"new_qcircuit.{name}({param1}, {param2}, {param3}, {param4}, {qubit_operand_1}, {qubit_operand_2})\n"
            f.write(gate_command)

        elif name == 'cu':
            gate_command = f"new_qcircuit.{name}({param1}, {param2}, {param3}, {param4}, {qubit_operand_1}, {qubit_operand_2})\n"
            f.write(gate_command)
    
        #elif name == 'cu1':
        #    f.write("new_qcircuit.cu1("+str(param1)+", "+str(qubit_operand_1)+", "+str(qubit_operand_2)+") \n")
    
        #need to debug CU, because it was deprecated, however, the RandomCircuit still assigns gates with that name. 
        
       # if name == 'xx_minus_yy':
       #     new_qcircuit.xx_minus_yy(param1, param2, qubit_operand_1, qubit_operand_2)
            
            
       #  if name == 'xx_plus_yy':
       #      new_qcircuit.xx_plus_yy(param1, param2, qubit_operand_1, qubit_operand_2)
        
       # need opflow   
                    

        #4 qubit gates (max operands for randomcircuit function)
        
        #f.close()
        
    f.write("new_qcircuit.draw()")
        
    print("Done.")
```
